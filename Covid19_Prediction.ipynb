```python
# importing pandas module to read csv file.
import pandas
```

```python
# reading data files using pandas 
# data file name is "data.csv"
Data = pandas.read_csv("data.csv", sep=",")
```

```python
# Display output of data file
Data
```

```python
# seperating data infected with Corona
DataofInfectedPeople = Data[Data[Data.columns[0]] == "Yes"]
```

```python
# Displaying output of infected people
DataofInfectedPeople
```

```python
# seperating data not infected with Corona
DataofNonInfectedPeople = Data[Data[Data.columns[0]] == "No"]
```

```python
# Displaying output of infected people
DataofNonInfectedPeople
```

```python
# input of symptoms by user
symptoms  = {'Dry Cough': 0.25, 'High Fever': 0.25, 'Sore Throat':0.25, 'Difficulty in breathing': 1}
```

```python
# Declaring dictionaries to store data

# This dictionary with store the count of symptoms of data with infected people 
CountsofSymptomswithInfected = {}

# This dictionary with store the count of symptoms of data with non infected people
CountsofSymptomswithNonInfected = {}
```

```python
# Calculating the count of symptoms in data of infected and non infected people
for key, data in symptoms.items():
    # Gives the count of key from data of infected people
    count = list(DataofInfectedPeople[key]).count(data)
    CountsofSymptomswithInfected[key] = count
    
    # Gives the count of key from data of non infected people
    count = list(DataofNonInfectedPeople[key]).count(data)
    CountsofSymptomswithNonInfected[key] = count
```

```python
# Displaying values in dictionary of infected people
CountsofSymptomswithInfected
```

```python
# Displaying values in dictionary of non infected people
CountsofSymptomswithNonInfected
```

```python
# The list will store the probabilities of infected people
AllProbabilitiesWithInfected =  []

# The list will store the probabilities of non infected people
AllProbabilityWithNonInfected =  []
```

```python
# total number of peoples we have in overall data
TotalNumberofPeople = len(list(Data['infected with covid']))
```

```python
# total number of peoples infected with Covid
NoofPeopleInfected = len(list(DataofInfectedPeople['infected with covid']))

# Probability of people infected with Corona
ProbabilityOfPeopleInfected = NoofPeopleInfected/TotalNumberofPeople
ProbabilityOfPeopleInfected
```

```python
# Appending ProbabilityOfPeopleInfected to list "AllProbabilitiesWithInfected"
AllProbabilitiesWithInfected.append(ProbabilityOfPeopleInfected)

# Calculating the probaility of symptoms given that you are infected with Corona
for key, data in CountsofSymptomswithInfected.items():
    AllProbabilitiesWithInfected.append(data/NoofPeopleInfected)
    
# Displaying the probaility given that you are infected with Corona
AllProbabilitiesWithInfected
```

```python
# total number of peoples not infected with Covid
NoofPeopleNotInfected = len(list(DataofNonInfectedPeople['infected with covid']))

# Probability of people not infected with Corona
ProbabilityOfPeopleNotInfected = NoofPeopleNotInfected/TotalNumberofPeople
ProbabilityOfPeopleNotInfected
```

```python
# Appending ProbabilityOfPeopleNotInfected to list "AllProbabilityWithNonInfected"
AllProbabilityWithNonInfected.append(ProbabilityOfPeopleNotInfected)

# Calculating the probaility of symptoms given that you are not infected with Corona
for key, data in CountsofSymptomswithNonInfected.items():
    AllProbabilityWithNonInfected.append(data/NoofPeopleNotInfected)
    
# Displaying the probaility given that you are not infected with Corona
AllProbabilityWithNonInfected
```

```python
# Multiplying the probailities given that you are infected with Corona
ProbabilityYouAreInfected = 1.0
for i in AllProbabilitiesWithInfected:
    if (i == 0): # if probaility of any symptom is equal to zero
        i = 0.01   # replace it with 0.01
    ProbabilityYouAreInfected = ProbabilityYouAreInfected*i
ProbabilityYouAreInfected
```

```python
# Multiplying the probailities given that you are not infected with Corona
ProbabilityYouAreNotInfected = 1.0
for i in AllProbabilityWithNonInfected:
    if (i == 0): # if probaility of any symptom is equal to zero
        i = 0.01   # replace it with 0.01
    ProbabilityYouAreNotInfected = ProbabilityYouAreNotInfected*i
ProbabilityYouAreNotInfected
```

```python
# Noramlize the probabilities 
TotalProbability = ProbabilityYouAreInfected + ProbabilityYouAreNotInfected
NProbabilityYouAreInfected = ProbabilityYouAreInfected/TotalProbability
NProbabilityYouAreNotInfected = ProbabilityYouAreNotInfected/TotalProbability
```

```python
# Displaying the Final result
if (NProbabilityYouAreInfected > NProbabilityYouAreNotInfected):
    print("You are infected with Corona.")

else:
    print("You are not infected with Corona.")
```
